## Simulação - moeda

Vamos simular a distribuição de eventos dada uma probabilidade predefinida

```python
import numpy as np
import pandas as pd

# estabelecemos qual a probabilidade que usaremos na simulacao
PROBABILITY = 0.6

# numero de simulacoes a serem realizadas
N_SIMULATIONS_ARRAY = [10, 100, 200, 500, 700, 1000, 3000, 5000, 7000, 10000]

# Lista com as saídas da moeda (Cara:1, Coroa:0)
simul_output = []

def generate_head_count(n_simul, prob):
    # contador de quantas vezes saiu cara
    head_count = 0
    for i in range(0, n_simul):
        if (np.random.uniform() < prob):
            head_count += 1
            simul_output.append(1)
        else:
            simul_output.append(0)
    return head_count / n_simul

df = pd.DataFrame(columns=['n_simulations', 'probability'])
for n in N_SIMULATIONS_ARRAY: 
    df = df.append(pd.DataFrame([[n, generate_head_count(n, PROBABILITY)]], columns=['n_simulations', 'probability']))

print(df)

df.plot(x='n_simulations', y='probability')
```

## Abordagem frequentista

```python
import numpy as np
import matplotlib.pyplot as plt

PROBABILITY = 0.6
prob_array = []
n_simul_array = []
N_MAX_SIMULATIONS = 10000

for n_simul in np.arange(1, N_MAX_SIMULATIONS, 10):
    head_count = 0
    for i in range(1, n_simul):
        if (np.random.uniform() < PROBABILITY):
            head_count += 1
    prob_array.append(head_count / n_simul)
    n_simul_array.append(n_simul)

plt.figure(figsize=(8, 6))
plt.plot(n_simul_array, prob_array, color='b', label='Valor simulado')
plt.axhline(y=PROBABILITY, color='r', label='Valor teorico')
plt.ylabel('Fracao de caras')
plt.xlabel('Numero de experimentos')
plt.xlim([0.0, N_MAX_SIMULATIONS])
plt.ylim([0.0, 1.0])
plt.legend()
plt.show(True)
```
